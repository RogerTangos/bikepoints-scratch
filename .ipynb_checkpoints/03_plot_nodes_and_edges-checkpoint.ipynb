```python
import random
import networkx as nx
import folium as f
```

```python
g = nx.read_gpickle("02_cambridge-small.gpickle")
```

# Experiment

```python
sample_node_data = list(g.nodes().data())[0]
sample_node = sample_node_data[0]
```

```python
neighbors = list(g.neighbors(sample_node))
neighbors
```

# Map

```python
m = f.Map(location = [42.3611108,-71.1079923], zoom_start=16)
nodes_data_subset = g.nodes().data()


# add nodes to graph
for node_data in nodes_data_subset:
    node_id = node_data[1]["id"]
    central = node_data[1].get('central')
    lon,lat = node_data[0][0], node_data[0][1]  
    
    if central:
        m.add_child(f.Marker(location=[lat,lon],popup=node_id))
    else:
        m.add_child(f.CircleMarker(location=[lat, lon], popup=node_id, color="red", radius=2))
    
# add edges
for node_data in nodes_data_subset:
    lon,lat = node_data[0]

    neighbors = g.neighbors(node_data[0])
    for neighbor in neighbors:
        edge = [[lat, lon], [neighbor[1], neighbor[0]]]
        
        m.add_child(f.PolyLine(locations=edge, line_opacity=0.25, color="blue"))
        
m    
```
