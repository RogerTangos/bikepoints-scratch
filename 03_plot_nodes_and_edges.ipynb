# Map

```python

m = f.Map(location = [42.3611108,-71.1079923], zoom_start=16)
nodes_data_subset = g.nodes().data()


# add edges
for node_data in nodes_data_subset:

    lon,lat = node_data[1]['lon'], node_data[1]['lat']

    neighbors = g.neighbors(node_data[0])
    for neighbor in neighbors:
        edge = [[lat, lon], [g.nodes[neighbor]['lat'], g.nodes[neighbor]['lon']]]
        
        m.add_child(f.PolyLine(locations=edge, line_opacity=0.1, color="gray"))

        
# add nodes to graph
for node_data in nodes_data_subset:
    node_id = node_data[1]["id"]
    voterank = node_data[1].get('voterank')
    betweenness = node_data[1].get('betweenness')
    lon,lat = node_data[1]['lon'], node_data[1]['lat']  
    
#     # distinguish between betweenness, voterank, and non-central nodes
#     if betweenness:
# #         m.add_child(f.Marker(location=[lat,lon],popup=node_id))
#         m.add_child(f.CircleMarker(location=[lat, lon], popup=node_id, color="blue", radius=5))
#     elif voterank:
#         m.add_child(f.CircleMarker(location=[lat, lon], popup=node_id, color="red", radius=4))
#     else:
#         m.add_child(f.CircleMarker(location=[lat, lon], popup=node_id, color="gray", radius=2, opacity=0.2))
    
        
m    
```
